```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Load the training dataset
train_df = pd.read_csv("data/CW1_train.csv")

# Set plot style
sns.set(style="whitegrid")

#Plot Histograms for Numerical Features
train_df.hist(figsize=(16, 12), bins=30, edgecolor='black')
plt.suptitle('Feature Distributions in CW1_train Dataset')
plt.show()
```

```python

#Pairplot to Visualize Relationships
sample_features = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z', 'outcome']
sns.pairplot(train_df[sample_features], diag_kind='kde', plot_kws={'alpha': 0.5})
plt.suptitle('Pairplot of Key Features')
plt.show()
```
